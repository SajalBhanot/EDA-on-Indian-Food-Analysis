# Data Manipulation

```python
flavour_counts = data['flavor_profile'].value_counts()
flavour_counts
```

```python
mycolors = ['orange','magenta','green','purple']
```

# Data Visualization

```python
# Plotting the pie chart
plt.figure(figsize=(5, 6))
plt.pie(flavour_counts, labels=flavour_counts.index, autopct='%1.1f%%', colors = mycolors, startangle=20)
plt.title('Distribution of Various Flavors of Dishes')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
diet_counts = data['diet'].value_counts()
```

```python
# Plotting the pie chart
plt.figure(figsize=(5, 6))
plt.pie(diet_counts, labels=diet_counts.index, autopct='%1.1f%%', colors = (['green','red']))
plt.title('Proportion of Vegetarian and Non-Vegetarian Dishes')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
course_counts = data['course'].value_counts()
course_counts
```

```python
# Plotting the pie chart
plt.figure(figsize=(5, 6))
plt.pie(course_counts, labels=course_counts.index, autopct='%1.1f%%', 
        colors = (['yellow','brown','orange','skyblue']), startangle=20)
plt.title('Course Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
region_counts = data['region'].value_counts()
region_counts
```

```python
fig, ax = plt.subplots()
ax.pie(region_counts, labels=region_counts.index, autopct='%1.1f%%', startangle=90)

# Draw a circle at the center of pie to make it a doughnut chart
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal')  

plt.title('Distribution By Region')
plt.show()
```

```python
state_counts = data['state'].value_counts()
state_counts
```

```python
yo = ['red', 'black', 'brown', 'orange', 'blue', 'yellow', 'green', 'pink', 'crimson', 'magenta', 'lightgreen', 'purple', 'darkblue', 'skyblue', 'cyan', 'grey']
```

```python
plt.figure(figsize=(10, 6))
state_counts.plot(kind='bar', color = yo)
plt.title('State Distribution')
plt.xlabel('State')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability if needed
plt.grid(axis='y', linestyle='-', alpha=0.7, color='black')
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
df = data[['name','prep_time']]
df.head()
```

```python
df = df.sort_values(by=['prep_time'],ascending=False)
df.head(10)
```

```python
# Line PLot
plt.plot(list(df['name'])[0:10],
        list(df['prep_time'])[0:10],
        color='black',linestyle='-',linewidth=3, marker='o', mfc = 'gold')
plt.title('Top 10 Dishes by Preparation Time')
plt.xlabel('Dishes Name')
plt.ylabel('Preparation Time')
plt.xticks(rotation=45, color='black')
plt.legend(['Preparation Time'])
plt.grid(True, color='grey')
plt.show()
```

```python
df = data[['name','cook_time']]
df.head()
```

```python
df = df.sort_values(by=['cook_time'],ascending=False)
df.head(10)
```
